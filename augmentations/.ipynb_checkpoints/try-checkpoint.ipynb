```python
from nltk.corpus import wordnet
import nltk
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag, word_tokenize
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
import spacy
import json
from pprint import pprint
nlp = spacy.load('en')
import textacy
import csv
import pandas as pd
import json
```

```python
def get_VP(sentence):
    pattern = r'<VERB>?<ADV>*<VERB>+'
    doc = textacy.Doc(sentence, lang='en')
    lists = textacy.extract.pos_regex_matches(doc, pattern)
    return lists
```

```python
def get_triplet(sentence):
    doc = nlp(sentence)
    triplet = {}
    for chunk in doc.noun_chunks:
        #print(chunk.text, chunk.root.text, chunk.root.dep_)
              #chunk.root.head.text)
        triplet[chunk.root.dep_] = chunk.text
    #print(triplet)
    return triplet
```

```python
f = open('store_triplets.json','w')
```

```python
temp = []
for question in data['questions'][:1000]:
    sentence = question['question']
    triplet = {}
    triplet['sentence'] = sentence
    attrs = get_triplet(sentence)
    triplet = {**triplet, **attrs}

    
    VPs = get_VP(sentence)

    if (triplet is not None) and (VPs is not None):
        verbs = []
        for vp in VPs:
            verbs.append(vp.text)
        triplet['verb phrase'] = verbs
        tagged = pos_tag(word_tokenize(sentence))
        verbs = []
#         for token in tagged:
#             if token[1].startswith('V'):
#                 verbs.append(token[0])
#         triplet.append(verbs)
        #json.dump(triplet,f)
        temp.append(triplet)
json.dump(temp,f)
```
